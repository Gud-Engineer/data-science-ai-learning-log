```python
# Write your code here

# np.tile(A, n): repeats array-like A, n times (concatenation).
# np.arange(L): creates integers [0..L-1] used to index cyclically via modulo.

# TODO Let n=2
n = 2
pattern = np.array([1, 2, 3])
L = 3*n
array_2 = pattern[np.arange(L) % len(pattern)]
array_2

# array_2alt = np.tile(pattern, n) # Pattern has 3 elements hence length is 3n
```

## Problem: 5

Create an 8 x 8 array with a checkerboard pattern of zeros and ones.

```python
# Write your code here

# Cell (i, j) should alternate based on parity of i+j:
# (i + j) even -> 0
# (i + j) odd  -> 1
# np.indices((n, n)) returns index grids for rows and columns:
# idx[0] is the row-index grid, idx[1] is the col-index grid.
# sum(axis=0) adds the row and column indices to form i+j over the grid & to that % 2 converts parity to 0/1.
checker_board = (np.indices((8, 8)).sum(axis=0) % 2).astype(int)
checker_board
```

## Problem: 7 
Create a function which creates an $n×n$ array with $(i , j)$ -entry equal to $i+ j$

```python
# Write your code here
def i_plus_j(n: int) -> np.ndarray:
    """Return an n x n array A where A[i, j] = i + j.

Approach : broadcasting :
- idx = [0, 1, ..., n-1]
- idx[:, None] shapes it as (n, 1) (a column vector)
- idx[None, :] shapes it as (1, n) (a row vector)
- Adding them broadcasts to (n, n), producing i+j for each cell.
"""
    idx = np.arange(n)
    return idx[:, None] + idx[None, :]

i_plus_j(5)

```

# Problem: 8
Evaluate $cos$ and $sin$ on the interval $[0,1]$  with 1000 points and then stack the results into a tall array with rows being the $(cos(x),sin(x))$ entries.

# Problem: 9

Create a random $3×5$ array using the `np.random.rand(3 ,5)` function and compute: the sum of all the entries, the sum of the rows, and the sum of the columns.

```python
# Write your code here

np.random.seed(42)  # for reproducible output
a = np.random.rand(3, 5)

min_val = a.min()
max_val = a.max()
mean_val = a.mean()
row_sums = a.sum(axis=1) 
col_sums = a.sum(axis=0)

print("Array:\n", a)
print("Min:", min_val)
print("Max:", max_val)
print("Mean:", mean_val)
print("Row sums:", row_sums)
print("Column sums:", col_sums)
```

## Problem: 10

Create a random $5×5$ array using the function `np.random.rand(5, 5)`. Then sort the rows according to the second column.

```python
# Write your code here

np.random.seed(42)
a = np.random.rand(5, 5)
order = a[:, 1].argsort()
sorted_a = a[order]

print("Original:\n", a)
print("\nSorted by 2nd column:\n", sorted_a)
```
