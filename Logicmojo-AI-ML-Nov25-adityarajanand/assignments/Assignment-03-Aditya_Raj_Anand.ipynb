#### 6. What are the five most common County names in the U.S.?

```python
# Write your code here
(
    pop["County"]                           # Select "County" column/ series object
    .value_counts()                         # Counts frequency of each "county" name
    .head(5)                                # Top 5
    .reset_index(name="Frequency")          # Converts the Series into a DataFrame, resetting the index, old index becomes column.
    .rename(columns={"index": "County"})    # Renames the column "index" to "County" to make it readable and intuitive.
)
```

#### 7. What are the top 5 most populated Counties according to the 2010 Census?

```python
# Write your code here
top5_county_2010 = (
    pop[["County", "2010Census"]]                           # Relevant columns
    .sort_values(by="2010Census", ascending=False)          # Sort Descending-Populated
    .head(5)                                                # Top 5
    .reset_index(drop=True)                                 # Clean Index
    .rename(columns={"2010Census": "Population_2010"})      # readable columns
)

top5_county_2010
```

#### 8. What are the top 5 most populated States according to the 2010 Census?

```python
# Write your code here

# Step 1: Aggregate 2010 populations per state
state_pop_2010 = pop.groupby("State")["2010Census"].sum().reset_index()
# Step 2: Sort by 2010 population in descending order
top5_states_2010 = (
    state_pop_2010
    .sort_values(by="2010Census", ascending=False)
    .head(5)
    .reset_index(drop=True)
    .rename(columns={"2010Census": "Population_2010"})
)

top5_states_2010
```

#### 9. How many Counties have 2010 populations greater than 1 million?

```python
# Write your code here
num_counties_over_million = len(pop[pop["2010Census"] > 1_000_000])
num_counties_over_million
```

#### 10. How many Counties don't have the word "County" in their name?

```python
# Write your code here

num_without_county = (~pop["County"].str.contains("County", case=False)).sum() # Use of ~ to negate search with County
num_without_county

```

#### 11. Add a column that calculates the percent change between the 2010 Census and the 2017 Population Estimate

```python
# Write your code here

pop["Percent_Change"] = ((pop["2017PopEstimate"] - pop["2010Census"]) / pop["2010Census"] * 100).round(2)
pop.head()

```

#### 12. What States have the highest estimated percent change between the 2010 Census and the 2017 Population Estimate?

**NOTE:** This will take several lines of code, as it requires a recalculation of Percent Change.

```python
# Write your code here

# Want to find State-Level Population Growth, per dataset is per County.
# Each row is county, and a state has multiple counties.

# Step 1 : Aggregate population per state : Sum county populations to get state totals
state_pop = pop.groupby("State")[["2010Census", "2017PopEstimate"]].sum().reset_index()
# state_pop.head() # for inspection

# Step 2: Calculate percent change per state
state_pop["Percent_Change"] = ((state_pop["2017PopEstimate"] - state_pop["2010Census"]) 
                               / state_pop["2010Census"] * 100).round(2)
# state_pop.head() # for inspection

# Step 3: Sort states by percent change (descending)
top_states_growth = state_pop.sort_values(by="Percent_Change", ascending=False).reset_index(drop=True)
top_states_growth.head(10)  # Show top 10 for inspection

# Step 4: Top 5 states with highest population growth
top5_states_growth = top_states_growth.head(5).rename(columns={"Percent_Change": "Percent_Change_2010_2017"})
top5_states_growth

```
