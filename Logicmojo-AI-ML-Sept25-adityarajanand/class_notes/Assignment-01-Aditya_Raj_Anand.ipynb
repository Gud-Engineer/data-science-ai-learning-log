````python
# Write your code here

def second_largest(numbers:list)->int | float:
    '''          
    Finds and returns the second largest number in a list of numbers.
    The function assumes that the input list contains at least two unique numbers.
    Duplicate numbers in the list are ignored when determining the second largest.

    Parameters:
    numbers (list): A list of integers or floats.

    Returns:
    int: The second largest number in the list.

    Example:
    >>> second_largest([10, 5, 20, 8, 15])
    15
    '''
    
    # result = sorted(list(set(numbers)),reverse=True)[1]
    # 1) remove duplicates (unique values only)
    unique_values = set(numbers)

    # 2) sort descending and pick the 2nd value
    #    (This assumes at least two unique numbers exist, as per the question)
    sorted_unique = sorted(unique_values, reverse=True)
    return sorted_unique[1]


if __name__ =="__main__":           
    # Test cases (similar style to Q2: list of inputs)
    test_cases = [
        [10, 5, 20, 8, 15],      # normal case
        [5, 4, 3, 2, 1],         # already sorted descending
        [10, 10, 20, 20, 30],    # duplicates present
        [1, 2],                  # smallest valid input (2 unique values)
        [100, 50, 100, 50],      # only 2 unique values repeated
        [-10, -20, -30, -40],    # negatives
        [0, 0, 1, 2],            # zeros + positives
        [1.5, 2.5, 3.5, 4.5],    # floats
        [1000000, 999999, 1],    # large numbers
    ]
      
    for idx, test_case in enumerate(test_cases):
        try:
            result = second_largest(test_case)
            print(f"TestCase {idx + 1}: {test_case} => Second Largest: {result}")
        except Exception as e:
            print(f"TestCase {idx + 1}: {test_case} => Error: {e}")
````

Example:
```python

```


````python
# Write your code here
def count_vowels(text: str) -> int:
    """
    Counts the number of vowels (a, e, i, o, u) in a given string.

    The function is case-insensitive and considers both uppercase and lowercase vowels.

    Parameters:
    text (str): The input string.

    Returns:
    int: The count of vowels in the string.

    Example:
    >>> count_vowels("Hello World")
    3    
    """
    vowels = set("aeiou")
    return sum(1 for char in text.lower() if char in vowels)


if __name__ == "__main__":    
    # Test cases (same style as Q1)
    test_cases = [
        "Hello World",           # mixed letters
        "PYTHON",                # uppercase only
        "aeiouAEIOU",            # all vowels, mixed case
        "This is a test case",   # multiple vowels
        "12345",                 # no letters
        "!!!@@@",                # only symbols
        "Mixed123Case!",         # letters + digits + symbols
        "A quick brown fox",     # sentence
        "",                      # empty string
        "E",                     # single vowel
        "bcdfghjklmnpqrstvwxyz"  # no vowels
    ]

    for idx, test_case in enumerate(test_cases):
        result = count_vowels(test_case)
        print(f"TestCase {idx + 1}: '{test_case}' => Vowel Count: {result}")
````

#### Question-3

Write a Python program to print all the prime numbers between 1 and N (inclusive). Where N is a positive integer and an input from user.

Example:

```python
N = 20
# Expected output: 2,3,5,7,11,13,17,19

```

````python
def is_prime(num: int) -> bool:
    """
    Determines whether a given number is a prime number.

    Parameters:
    num (int): The number to check for primality.

    Returns:
    bool: True if the number is prime, False otherwise.
    """
    if num <= 1:
        return False
    
    # Try dividing by every integer from 2 up to sqrt(num)
    i = 2
    while i * i <= num: # i*i <= num means i <= sqrt(num) but avoids floating point
        if num % i == 0:
            return False
        i += 1
    return True


def print_primes_up_to_n(n: int) -> list:
    """
    Returns all prime numbers between 1 and n (inclusive).

    Parameters:
    n (int): The upper limit (inclusive) to find prime numbers.

    Returns:
    list: A list of all prime numbers between 1 and n.
    """
    if n < 2:
        return []
    primes = [num for num in range(2, n + 1) if is_prime(num)]
    return primes


if __name__ == "__main__":
    # Input from user
    try:
        N = int(input("Enter a positive integer N: "))
        if N < 1:
            print("Please enter a positive integer greater than 0.")
        else:
            primes = print_primes_up_to_n(N)
            # Format the output as a comma-separated string
            print(f"Prime numbers between 1 and {N}: {', '.join(map(str, primes))}")    
    except ValueError:
        print("Invalid input! Please enter a valid positive integer.")
        print("Executing Test Cases instead")         
        
    # ---- Additional test cases (learning / validation) ----
    test_cases = [1, 2, 3, 10, 20, 30, 0, -5, 50]


    print("\nTest Cases:")
    for idx, test_case in enumerate(test_cases):
        if test_case < 1:
            print(f"TestCase {idx + 1}: N = {test_case} => Output: Invalid input")
        else:
            result = print_primes_up_to_n(test_case)
            print(f"TestCase {idx + 1}: N = {test_case} => Primes: {', '.join(map(str, result))}")

    
       
    
````

#### Question-4

Write a function `merge_sorted_lists(list1, list2)` that takes two sorted lists of integers, `list1` and `list2` and merges them into a single sorted list efficiently. One way is to simple concatenate the lists and sort. But think of a better approach.

Example:
```python
list1 = [1, 4, 5]
list2 = [1, 3, 4]
merged = merge_sorted_lists(list1, list2)
print(merged)
# Expected output: [1, 1, 3, 4, 4, 5]
```

````python
# Write your code here
def merge_sorted_lists(list1:list,list2:list)->list:
    """
    Merge two already-sorted lists into a single sorted list.

    This uses the two-pointer technique (like the merge step in merge sort).

    Args:
        list1: First sorted list
        list2: Second sorted list

    Returns:
        A new sorted list containing all elements from list1 and list2.

    Example:
        merge_sorted_lists([1,3,5], [2,4,6]) -> [1,2,3,4,5,6]
    """
    i, j = 0, 0
    merged = []
    
    # Compare current elements and take the smaller one each time
    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged.append(list1[i])
            i += 1
        else:
            merged.append(list2[j])
            j += 1

    # Append remaining elements (at most one of these will add anything)
    merged.extend(list1[i:])
    merged.extend(list2[j:])
    return merged

if __name__ == "__main__":
    print(merge_sorted_lists([1, 3, 5], [2, 4, 6]))     # [1,2,3,4,5,6]
    print(merge_sorted_lists([], [1, 2]))               # [1,2]
    print(merge_sorted_lists([1, 2], []))               # [1,2]
    print(merge_sorted_lists([1, 2, 2], [2, 3]))        # [1,2,2,2,3]
````

````python
# Write your code here

def find_highest_grade(grades: dict[str, int]) -> str:
    """
    Return the name of the student with the highest grade.

    Args:
        grades: dict mapping student name -> grade

    Returns:
        Name of the student with the highest grade.

    Raises:
        ValueError: if grades is empty.
    """
    if not grades:
        raise ValueError("grades dictionary is empty")

    return max(grades, key=grades.get)


if __name__ == "__main__":
    student_grades = {"Alice": 85, "Bob": 92, "Charlie": 78}
    print(find_highest_grade(student_grades))  # Bob

    # Additional test cases
    test_cases = [
        ({"Alice": 85, "Bob": 92, "Charlie": 78}, "Bob"),
        ({"X": 10, "Y": 10, "Z": 5}, "X"),  # tie -> first inserted
        ({"OnlyOne": 100}, "OnlyOne"),
        ({"A": -1, "B": -5, "C": -3}, "A"), # works with negative grades too
    ]

    print("\nTest Cases:")
    for idx, (grades, expected) in enumerate(test_cases, start=1):
        try:
            result = find_highest_grade(grades)
            print(f"TestCase {idx}: {grades} => {result} | Expected: {expected}")
        except Exception as e:
            print(f"TestCase {idx}: {grades} => Error: {e}")

    # Empty dict edge case (should raise ValueError)
    try:
        find_highest_grade({})
    except ValueError as e:
        print(f"EdgeCase (empty dict) => Correctly raised: {e}")

# #-----------------#
# methods = [name for name in dir(student_grades)
#            if callable(getattr(student_grades, name)) and not name.startswith("__")]
# print(methods)
# #-----------------#
# help(student_grades.get)


````

#### Question-6

Write a function `is_anagram(str1, str2)` that checks if two given strings are anagrams of each other (contain the same characters with the same frequencies, ignoring case).

Example:
```python
print(is_anagram("listen", "silent"))
print(is_anagram("hello", "world"))
# Expected output:
# True
# False
```

````python
# Write your code here

# word1 = "listen"
# word2 = "silent"

# # I need a dictionary with letters and fequency -> There is a pythonic way to do it !!
# from collections import Counter

# Counter(word1)

from collections import Counter
def is_anagram(word1:str,word2:str)->bool:
    """    
    Check if two strings are anagrams (case-insensitive).

    Anagrams: same letters with same counts, order doesn't matter.

    Args:
        str1: first string
        str2: second string

    Returns:
        True if they are anagrams, else False.

    Example:
        is_anagram("listen", "silent") -> True
        is_anagram("hello", "world") -> False
    """   
    return Counter(word1.lower()) == Counter(word2.lower())

if __name__ == "__main__":
    print(is_anagram("listen", "silent"))  # True
    print(is_anagram("hello", "world"))    # False

    # Additional test cases
    test_cases = [
        ("Listen", "Silent", True),          # case-insensitive
        ("aabbcc", "abcabc", True),          # repeated letters
        ("rat", "car", False),               # different letters
        ("", "", True),                      # both empty
        ("A", "a", True),                    # single char, case-insensitive
    ]

    print("\nTest Cases:")
    for idx, (a, b, expected) in enumerate(test_cases, start=1):
        result = is_anagram(a, b)
        print(f"TestCase {idx}: '{a}' vs '{b}' => {result} | Expected: {expected}")
````

#### Question-7

Write a function `rotate_list(lst, k)` that rotates a given list lst to the right by `k` positions in-place. Assume `k` is a non-negative integer and can be greater than the length of the list.

Example:

```python
my_list = [1, 2, 3, 4, 5]
rotate_list(my_list, 2)
print(my_list)
# Expected output: [4, 5, 1, 2, 3]
```

````python
# # Write your code here
def rotate_list(lst:list[int], k:int) -> list[int]:
    """
    Rotate a list to the right by k steps, in-place.

    Args:
        lst: list to rotate (modified in place)
        k: number of steps to rotate right

    Returns:
        The same list object after rotation.

    Example:
        lst = [1,2,3,4,5]
        rotate_list(lst, 2) -> [4,5,1,2,3]
    """
    n = len(lst)
    if n == 0:
        return lst  # nothing to rotate original list

    k = k % n  # handle k > n : 
    if k == 0:
        return lst  # # rotation by 0 (or multiple of n) leaves list unchanged

    # Build rotated version, then write back into lst in-place
    rotated = lst[-k:] + lst[:-k]
    lst[:] = rotated  # in-place modification (same list object)
    return lst

# my_list = [1, 2, 3, 4, 5]
# print(rotate_list(my_list,2))    
# print(my_list) # Just checking : In-place or not !! 
if __name__ == "__main__":
    a = [1, 2, 3, 4, 5]
    print(rotate_list(a, 2))  # [4, 5, 1, 2, 3]
    print(a)                  # [4, 5, 1, 2, 3]

    # Additional test cases
    test_cases = [
        {"lst": [1, 2, 3, 4, 5], "k": 0,  "expected": [1, 2, 3, 4, 5]},
        {"lst": [1, 2, 3, 4, 5], "k": 5,  "expected": [1, 2, 3, 4, 5]},  # k == n
        {"lst": [1, 2, 3, 4, 5], "k": 7,  "expected": [4, 5, 1, 2, 3]},  # k > n
        {"lst": [1],             "k": 10, "expected": [1]},              # single element
        {"lst": [],              "k": 3,  "expected": []},               # empty list
        {"lst": [10, 20, 30],    "k": -1, "expected": [20, 30, 10]},     # negative k (works via modulo)
    ]

    print("\nTest Cases:")
    for idx, case in enumerate(test_cases, start=1):
        arr = case["lst"].copy()
        rotate_list(arr, case["k"])
        print(f"TestCase {idx}: lst={case['lst']}, k={case['k']} => {arr} | Expected: {case['expected']}")
````

You have a list of dictionaries, where each dictionary represents a product with keys "name", "price", and "category". Write a function group_by_category(products) that takes this list and returns a dictionary where the keys are the unique categories and the values are lists of product names belonging to that category.

Example:

```python
products = [
    {"name": "Laptop", "price": 1200, "category": "Electronics"},
    {"name": "Book", "price": 25, "category": "Books"},
    {"name": "Tablet", "price": 300, "category": "Electronics"},
    {"name": "Fiction", "price": 15, "category": "Books"},
]
grouped_products = group_by_category(products)
print(grouped_products)
# Expected output: {'Electronics': ['Laptop', 'Tablet'], 'Books': ['Book', 'Fiction']}

```


````python
# Write your code here

def  group_by_category(product: list[dict]):
    """
    Group product names by their category.

    Input:
        products: list of dicts, each dict contains:
          - 'name'
          - 'price'
          - 'category'

    Output:
        dict where:
          - key   = category (string)
          - value = list of product names in that category

    Example:
        [{'name':'Laptop','category':'Electronics'}, ...]
        -> {'Electronics': ['Laptop', ...], ...}
    """
    result: dict[str, list[str]] = {}

    for product in products:
        category = product["category"]
        name = product["name"]

        # If this category is not present, start a new list
        if category not in result:
            result[category]= []
        
        result[category].append(name)

    return result

if __name__ == "__main__":
    # Example from the question
    products = [
        {"name": "Laptop", "price": 1200, "category": "Electronics"},
        {"name": "Book", "price": 25, "category": "Books"},
        {"name": "Tablet", "price": 300, "category": "Electronics"},
        {"name": "Fiction", "price": 15, "category": "Books"},
    ]
    grouped_products = group_by_category(products)
    print(grouped_products)
    # Expected: {'Electronics': ['Laptop', 'Tablet'], 'Books': ['Book', 'Fiction']}

    # Additional test cases
    test_cases = [
        (
            products,
            {'Electronics': ['Laptop', 'Tablet'], 'Books': ['Book', 'Fiction']}
        ),
        (
            [{"name": "Pen", "price": 2, "category": "Stationery"}],
            {"Stationery": ["Pen"]}
        ),
        (
            [],
            {}
        ),
        (
            [
                {"name": "A", "price": 1, "category": "C1"},
                {"name": "B", "price": 2, "category": "C1"},
                {"name": "C", "price": 3, "category": "C2"},
            ],
            {"C1": ["A", "B"], "C2": ["C"]}
        ),
    ]

    print("\nTest Cases:")
    for idx, (inp, expected) in enumerate(test_cases, start=1):
        result = group_by_category(inp)
        print(f"TestCase {idx}: {result} | Expected: {expected}")
````

#### Question-9

Write a Python program that prints the first `n` rows of Pascal's triangle. `n` is a positive integer and an user input. To know about Pascal's triangle see the [wikipedia article](https://en.wikipedia.org/wiki/Pascal%27s_triangle): 

Example:

```python
n = 5
# Expected output:
#     1
#    1 1
#   1 2 1
#  1 3 3 1
# 1 4 6 4 1
```

````python
# Write your code here
def generate_pascal_triangle(n):
    """
    Generate the first n rows of Pascal's triangle.

    Args:
        n: number of rows (positive integer)

    Returns:
        A list of rows, where each row is a list of ints.

    Example:
        generate_pascal_triangle(3) -> [[1], [1, 1], [1, 2, 1]]
    """
    if n <= 0:
        return []
    
    triangle: list[list[int]] = []

    for r in range(n):
        if r == 0:
            # First row is always [1]
            triangle.append([1])
        else:
            prev = triangle[-1]       # previous row
            row = [1]                 # first element is always 1

            # Compute inner elements
            for i in range(1, len(prev)):
                row.append(prev[i - 1] + prev[i])

            row.append(1)             # last element is always 1
            triangle.append(row)

    return triangle


def print_pascal_triangle(n):
    """
    Print the first n rows of Pascal's triangle in a centered format.
    """
    triangle = generate_pascal_triangle(n)

    if not triangle:
        return

    # Compute width of the last row when printed as "1 4 6 4 1"
    last_row_str = " ".join(str(x) for x in triangle[-1])
    width = len(last_row_str)

    # Print each row centered within this width
    for row in triangle:
        row_str = " ".join(str(x) for x in row)
        print(row_str.center(width))


if __name__ == "__main__":
    # Example from the question
    n = 5
    print_pascal_triangle(n)

    # Additional test cases
    test_cases = [1, 2, 3, 5, 0, -2]
    print("\nTest Cases (generate_pascal_triangle):")
    for idx, tc in enumerate(test_cases, start=1):
        tri = generate_pascal_triangle(tc)
        print(f"TestCase {idx}: n={tc} => {tri}")
````

#### Question-10

Write a function `reverse_the_words(s)` which will take a string `s` as input and reverses the words in the string. 

Example:

```python
s = "Python is awesome and challenging"
r = reverse_the_words(s)
print(r)
# Expected output: challenging and awesome is Python
```

````python
# Write your code here
def reverse_the_words(s: str) -> str:
    """
    Reverse the order of words in a string.

    Args:
        s: input string

    Returns:
        A string with words in reverse order.

    Example:
        reverse_the_words("Python is awesome") -> "awesome is Python"
    """
    # return " ".join(s.split()[::-1]) # TODO : One liner !
    words = s.split()      # converts string -> list of words
    words.reverse()        # reverse the list in place
    return " ".join(words) # join back into a string


if __name__ == "__main__":
    # Example from the question
    s = "Python is awesome and challenging"
    print(reverse_the_words(s))  # Expected: challenging and awesome is Python

    # Additional test cases
    test_cases = [
        ("Hello World", "World Hello"),
        ("  multiple   spaces here ", "here spaces multiple"),
        ("One", "One"),
        ("", ""),
        ("a b c", "c b a"),
    ]

    print("\nTest Cases:")
    for idx, (inp, expected) in enumerate(test_cases, start=1):
        result = reverse_the_words(inp)
        print(f"TestCase {idx}: '{inp}' => '{result}' | Expected: '{expected}'")
````

#### Bonus question

Write a function `word_counter(s)` which will take a string `s` as input and returns a dictionary with count of words (case insensitive). Remove all the punctuations and special symbols present in the string.  

Example:
```python
s = "This city is a beautiful city. This is where I live"
d = word_counter(s)

# Expected output (output can be in any order): {"this": 2, "city": 2, "is": 2, "a": 1, "beautiful": 1, "where": 1, "i": 1, "live": 1}
```

````python
# Write your code here
from collections import Counter
import re

def word_counter(s:list)-> dict:
    """
    Count word frequency in a string (case-insensitive), ignoring punctuation.

    A "word" here means consecutive letters/digits (A-Z, a-z, 0-9).

    Args:
        text: input string

    Returns:
        dict mapping word -> count

    Example:
        word_counter("Hi hi!") -> {"hi": 2}
    """
    # s = re.sub(r"[^A-Za-z0-9\s]+", "", s.lower())  # Cleaning S. Keep letters/digits/whitespace
    # words_1 = s.split(" ")    
    
    # words_2 =  re.split(r"\W+", s.lower()) # Split on anything that not [a-zA-Z0-9_]

    words = re.findall(r"[A-Za-z0-9]+", s.lower())


    result = Counter(words)
    # print(result)
    return result

if __name__ == "__main__":
    # Example from the question
    s = "This city is a beautiful city. This is where I live"
    print(word_counter(s))

    # Additional test cases
    test_cases = [
        ("Hi hi!", {"hi": 2}),
        ("Numbers 123 123", {"numbers": 1, "123": 2}),
        ("Punctuation!!! should; be: ignored.", {"punctuation": 1, "should": 1, "be": 1, "ignored": 1}),
        ("", {}),
        ("One-word", {"one": 1, "word": 1}),  # hyphen splits into two tokens by our regex rule
    ]

    print("\nTest Cases:")
    for idx, (text, expected) in enumerate(test_cases, start=1):
        result = word_counter(text)
        print(f"TestCase {idx}: '{text}' => {result} | Expected: {expected}")
````
